```python
#Import the necessary libraries
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your time series data
df = pd.read_csv('monthly_sales.csv', parse_dates=['date'], index_col='date')
```

```python
#Moving Averages
#Using moving averages to smooth the time series and identify trends.
# Calculate moving averages
df['moving_avg'] = df['value'].rolling(window=12).mean()

# Plot original series and moving average
plt.figure(figsize=(12, 6))
plt.plot(df['value'], label='Original')
plt.plot(df['moving_avg'], label='Moving Average', color='red')
plt.legend()
plt.show()
```
